```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset you just generated
df = pd.read_csv("data/processed/master_dataset_advanced_v2.csv")

# 2. Basic Inspection
print(f"Dataset Shape: {df.shape}")
print("-" * 30)
print("Columns Available:")
print(df.columns.tolist())

# 3. Check for Missing Values (Crucial for DML)
# DML models often fail if there are NaNs in the control variables
print("-" * 30)
print("Missing Values per Column:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# 4. Peek at the Top Earners (Sanity Check)
print("-" * 30)
print("Top 5 Salaries:")
print(df[['PLAYER_NAME', 'Salary']].sort_values(by='Salary', ascending=False).head())
```
